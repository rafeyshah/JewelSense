```python
import os
from shutil import copy2
from glob import glob

# Dataset paths
roboflow_path = "/content/ultralytics/Jewellery_detect-1"
dress_path = "/content/drive/MyDrive/dress_dataset"
output_path = "/content/drive/MyDrive/combined_dataset"

# Class mappings
roboflow_keep = {'ring': 7, 'earring': 3}
remap_roboflow = {7: 0, 3: 1}

dress_keep = {'dress': 0}
remap_dress = {0: 2}

# Helper function
def filter_and_remap_split(split_name, source_base, keep, remap, dst_base):
    src_img_dir = os.path.join(source_base, split_name, "images")
    src_lbl_dir = os.path.join(source_base, split_name, "labels")

    dst_img_dir = os.path.join(dst_base, split_name, "images")
    dst_lbl_dir = os.path.join(dst_base, split_name, "labels")

    os.makedirs(dst_img_dir, exist_ok=True)
    os.makedirs(dst_lbl_dir, exist_ok=True)

    for label_path in glob(os.path.join(src_lbl_dir, "*.txt")):
        with open(label_path, "r") as f:
            lines = f.readlines()

        new_lines = []
        for line in lines:
            parts = line.strip().split()
            cls_id = int(parts[0])
            if cls_id in keep.values():
                new_cls_id = remap[cls_id]
                new_line = " ".join([str(new_cls_id)] + parts[1:]) + "\n"
                new_lines.append(new_line)

        if new_lines:
            base = os.path.basename(label_path)
            new_lbl = os.path.join(dst_lbl_dir, base)
            with open(new_lbl, "w") as f_out:
                f_out.writelines(new_lines)

            img_path = os.path.join(src_img_dir, base.replace(".txt", ".jpg"))
            if os.path.exists(img_path):
                copy2(img_path, os.path.join(dst_img_dir, os.path.basename(img_path)))

# Splits to handle
splits = ["train", "valid", "test"]

# Merge Roboflow dataset
for split in splits:
    filter_and_remap_split(split, roboflow_path, roboflow_keep, remap_roboflow, output_path)

# Merge Dress dataset
for split in splits:
    filter_and_remap_split(split, dress_path, dress_keep, remap_dress, output_path)

# Write data.yaml
yaml_content = f"""
train: {output_path}/train/images
val: {output_path}/valid/images
test: {output_path}/test/images

nc: 3
names: ['ring', 'earring', 'dress']
"""

with open(os.path.join(output_path, "data.yaml"), "w") as f:
    f.write(yaml_content.strip())

print("âœ… Split-based combined dataset created.")
print("ğŸ‘‰ Saved at:", output_path)
```
